### DFS 알고리즘 구현
자료구조 스택과 큐를 활용함
- need_visit 스택과 visited 큐, 두 개의 자료 구조를 생성

> BFS 자료구조는 두 개의 큐를 활용하는데 반해, DFS는 스택과 큐를 활용한다

```python
def dfs(graph, start_node):
    visited,need_visit = list(), list()
    need_visit.append(start_node)

    while need_visit:
        node = need_visit.pop() # 뒤 뺌
        if node not in visited:
            visited.append(node)
            need_visit.extend(graph[node]) # 뒤 붙

    return visited
```

```python
dfs(graph, 'A')
```
