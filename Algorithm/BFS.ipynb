너비 우선 탐색(Breadth-First Search)

대표적인 그래프 탐색 알고리즘
- 너비 우선 탐섹 : 정점들과 같은 레벨에 있는 노드 (형제 노드들)을 먼저 탐색하는 방식
- 깊이 우선 탐색 : 정점의 자식들을 먼저 탐색하는 방식

```python
graph = dict()

graph['A'] = ['B', 'C']
graph['B'] = ['A', 'D']
graph['C'] = ['A', 'G', 'H', 'I']
graph['D'] = ['B', 'E', 'F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C']
graph['H'] = ['C']
graph['I'] = ['C', 'J']
graph['J'] = ['I']
```

```python
graph
```

알고리즘 구현

자료구조 큐를 활용함
- need_visit 큐와 visited 큐, 두 개의 큐를 생성

```python
def bfs(graph, root):
    visited = list()
    need_visit = list()

    need_visit.append(root)
    count = 0
    while need_visit:
        count += 1
        node = need_visit.pop(0)
        if node not in visited:
            visited.append(node)
            need_visit.extend(graph[node])
    print(count)
    return visited
```

```python
bfs(graph,'A')
```

시간 복잡도
- 노드 수 : V
- 간선 수 : E

시간 복잡도 : O(V+E)
